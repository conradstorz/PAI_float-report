```python
import pandas as pd
csvFile = pd.read_csv('t.csv', parse_dates=['Date'])
csvFile.head(10)
```

```python
print('Convert')
csvFile['Amount'] = csvFile['Amount'].astype(str)
csvFile.head(15)
```

```python
csvFile['Amount'] = csvFile['Amount'].str.lstrip('$')
csvFile['Amount'] = pd.to_numeric(csvFile['Amount'])
csvFile.groupby('Type')['Amount'].sum()
```

```python
csvFile['Amount'].where(~(csvFile.Type == 'Discount'), other=-csvFile['Amount'], inplace=True)
csvFile['Amount'].where(~(csvFile.Type == 'Payment'), other=-csvFile['Amount'], inplace=True)
csvFile['x-ponder'] = csvFile['Transponder'].astype(str) + csvFile['Plate'].astype(str)
csvFile['x-ponder'].where(~(csvFile['x-ponder'] == 'nannan'), other='Payments & Discounts', inplace=True)
csvFile['x-ponder'].where(~(csvFile['x-ponder'] == '6712755.0nan'), other='Equinox', inplace=True)
csvFile['x-ponder'].where(~(csvFile['x-ponder'] == 'nan825ZYS'), other='Equinox', inplace=True)
csvFile['x-ponder'].where(~(csvFile['x-ponder'] == '7020623.0nan'), other='Buick', inplace=True)
csvFile['x-ponder'].where(~(csvFile['x-ponder'] == 'nan358LJT'), other='Buick', inplace=True)
csvFile['x-ponder'].where(~(csvFile['x-ponder'] == '6665380.0nan'), other='Honda Fit?', inplace=True)
csvFile.head(10)
```

```python
csvFile.groupby(['x-ponder'], dropna=False).agg({'Amount': 'sum'})
```

```python
csvFile.groupby(['Date', 'x-ponder'], dropna=False).agg({'Amount': 'sum'})
```
